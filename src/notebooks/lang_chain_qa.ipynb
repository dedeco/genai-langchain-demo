### **O que √© LangChain?**
> LangChain √© uma estrutura para desenvolvimento de aplicativos alimentados por grandes modelos de linguagem (LLMs).

**TL;DR** LangChain facilita as partes complicadas de trabalhar e construir modelos de linguagem. Isso ajuda a fazer isso de duas maneiras:

1. **Integra√ß√µes** - Traga dados externos, como seus arquivos, outros aplicativos e dados de API, para LLMs
2. **Agentes** - Permite que os LLMs interajam com seu ambiente por meio da tomada de decis√µes e usem LLMs para ajudar a decidir qual a√ß√£o tomar a seguir

image.png

Para construir aplica√ß√µes eficazes de IA generativa, √© fundamental permitir que os LLMs interajam com sistemas externos. Isso torna os modelos conscientes dos dados e agentes, o que significa que eles podem compreender, raciocinar e usar os dados para agir de maneira significativa. Os sistemas externos podem ser corpus de dados p√∫blicos, reposit√≥rios de conhecimento privados, bancos de dados, aplicativos, APIs ou acesso √† Internet p√∫blica por meio da Pesquisa Google.


Aqui est√£o alguns padr√µes onde os LLMs podem ser aumentados com outros sistemas:
- Converter linguagem natural para SQL, executando o SQL no banco de dados, analisar e apresentar os resultados
- Chamar um webhook ou API externo com base na consulta do usu√°rio
- Sintetize resultados de v√°rios modelos ou encadeie os modelos em uma ordem espec√≠fica

Pode parecer trivial reunir essas chamadas e orquestr√°-las, mas torna-se uma tarefa mundana escrever c√≥digo de cola repetidamente, por exemplo. para cada conector de dados diferente ou um novo modelo. √â a√≠ que entra o LangChain!

image.png

### **Por que LangChain?**

A implementa√ß√£o modular de componentes e padr√µes comuns do LangChain que combina esses componentes facilita a constru√ß√£o de aplicativos complexos baseados em LLMs. LangChain permite que esses modelos se conectem a fontes de dados e sistemas como agentes para agir.


1. **Componentes** s√£o abstra√ß√µes que funcionam para trazer dados externos, como documentos, bancos de dados, aplicativos, APIs para modelos de linguagem. LangChain facilita a troca de abstra√ß√µes e componentes necess√°rios para trabalhar com LLMs.

2. **Agentes** permitir que os modelos de linguagem se comuniquem com seu ambiente, onde o modelo ent√£o decide a pr√≥xima a√ß√£o a ser tomada. LangChain fornece suporte pronto para uso para usar e personalizar 'cadeias' - uma s√©rie de a√ß√µes interligadas.

Embora os LLMs possam ser diretos (entrada e sa√≠da de texto), voc√™ rapidamente se deparar√° com pontos de atrito nos quais o LangChain ajuda quando voc√™ desenvolve aplicativos mais complicados.

Next you will define some utility functions that you will use for the Vertex AI Embeddings API

````python
# Utility functions for Embeddings API with rate limiting
def rate_limit(max_per_minute):
    period = 60 / max_per_minute
    print("Waiting")
    while True:
        before = time.time()
        yield
        after = time.time()
        elapsed = after - before
        sleep_time = max(0, period - elapsed)
        if sleep_time > 0:
            print(".", end="")
            time.sleep(sleep_time)


class CustomVertexAIEmbeddings(VertexAIEmbeddings, BaseModel):
    requests_per_minute: int
    num_instances_per_batch: int

    # Overriding embed_documents method
    def embed_documents(self, texts: List[str]):
        limiter = rate_limit(self.requests_per_minute)
        results = []
        docs = list(texts)

        while docs:
            # Working in batches because the API accepts maximum 5
            # documents per request to get embeddings
            head, docs = (
                docs[: self.num_instances_per_batch],
                docs[self.num_instances_per_batch :],
            )
            chunk = self.client.get_embeddings(head)
            results.extend(chunk)
            next(limiter)

        return [r.values for r in results]
````

# Componentes LangChain

Vamos fazer um r√°pido tour pela estrutura e conceitos do LangChain que voc√™ deve conhecer. LangChain oferece uma variedade de m√≥dulos que podem ser usados para criar aplicativos de modelo de linguagem. Esses m√≥dulos podem ser combinados para criar aplica√ß√µes mais complexas ou podem ser usados individualmente para aplica√ß√µes mais simples.

image.png

- **Modelos** s√£o o alicerce do LangChain, fornecendo uma interface para diferentes tipos de modelos de IA. Modelos de grandes idiomas (LLMs), modelos de bate-papo e incorpora√ß√µes de texto s√£o tipos de modelo suportados.
- **Prompts** referem-se √† entrada do modelo, que normalmente √© constru√≠do a partir de v√°rios componentes. LangChain fornece interfaces para construir e trabalhar facilmente com prompts - modelos de prompt, seletores de exemplo e analisadores de sa√≠da.
- **Mem√≥ria** fornece uma constru√ß√£o para armazenar e recuperar mensagens durante uma conversa que pode ser de curto ou longo prazo.
- **√çndices** ajudam os LLMs a interagir com documentos, fornecendo uma maneira de estrutur√°-los. LangChain fornece carregadores de documentos para carregar documentos, divisores de texto para dividir documentos em peda√ßos menores, armazenamentos de vetores para armazenar documentos como incorpora√ß√µes e recuperadores para buscar documentos relevantes.
- **Cadeias** permitem combinar componentes modulares (ou outras cadeias) em uma ordem espec√≠fica para concluir uma tarefa.
- **Agentes** s√£o uma constru√ß√£o poderosa no LangChain, permitindo que os LLMs se comuniquem com sistemas externos por meio de ferramentas e observem e decidam o melhor curso de a√ß√£o para concluir uma determinada tarefa.

### **Modelos de Prompt**

[Prompt Template](https://python.langchain.com/en/latest/modules/prompts/prompt_templates.html) √© um objeto que ajuda a criar prompts com base em uma combina√ß√£o de entrada do usu√°rio, outras informa√ß√µes n√£o est√°ticas e uma sequ√™ncia de modelo fixa.

### **Seletores de exemplo**

[Seletores de exemplo](https://python.langchain.com/en/latest/modules/prompts/example_selectors.html) s√£o uma maneira f√°cil de selecionar entre uma s√©rie de exemplos para colocar dinamicamente informa√ß√µes contextualizadas em seu prompt. Freq√ºentemente usado quando a tarefa tem nuances ou tem uma grande lista de exemplos.

Confira diferentes tipos de seletores de exemplo [aqui](https://python.langchain.com/docs/modules/model_io/prompts/example_selectors/)

````python
from langchain.prompts import FewShotPromptTemplate, PromptTemplate
from langchain.prompts.example_selector import SemanticSimilarityExampleSelector
from langchain.vectorstores import FAISS

example_prompt = PromptTemplate(
    input_variables=["input", "output"],
    template="Exemplo de entrada: {input}\nExemplo de sa√≠da: {output}",
)

# Exemplos de locais onde os substantivos s√£o encontrados
examples = [
    {"input": "pirata", "output": "navio"},
    {"input": "piloto", "output": "avi√£o"},
    {"input": "motorista", "output": "carro"},
    {"input": "√°rvore", "output": "ch√£o"},
    {"input": "p√°ssaro", "output": "ninho"},
]
````

````python
# SemanticSimilarityExampleSelector selecionar√° exemplos semelhantes √† sua entrada por significado sem√¢ntico

example_selector = SemanticSimilarityExampleSelector.from_examples(
    # Esta √© a lista de exemplos dispon√≠veis para sele√ß√£o.
    examples,
    # Esta √© a classe de incorpora√ß√£o usada para produzir embeddings que s√£o usados para medir a similaridade sem√¢ntica.
    embeddings,
    # Esta √© a classe VectorStore usada para armazenar os embeddings e fazer uma pesquisa de similaridade.
    FAISS,
    # Este √© o n√∫mero de exemplos a serem produzidos.
    k=2,
)
````

````python
similar_prompt = FewShotPromptTemplate(
    # O objeto que ajudar√° a selecionar exemplos
    example_selector=example_selector,
    # Seu prompt
    example_prompt=example_prompt,
    # Personaliza√ß√µes que ser√£o adicionadas na parte superior e inferior do seu prompt
    prefix="Forne√ßa a localiza√ß√£o em que um item geralmente √© encontrado",
    suffix="Input: {noun}\nOutput:",
    # Quais entradas seu prompt receber√°
    input_variables=["noun"],
)
````

````python
# Selecione um substantivo!
my_noun = "estudante"

print(similar_prompt.format(noun=my_noun))
````

````python
llm(similar_prompt.format(noun=my_noun))
````

### **Analisadores de sa√≠das**

[Output Parsers](https://python.langchain.com/en/latest/modules/prompts/output_parsers.html) ajuda a formatar a sa√≠da de um modelo. Geralmente usado para sa√≠da estruturada.

Duas ideias principais:

**1. Instru√ß√µes de formato**: um prompt gerado automaticamente que informa ao LLM como formatar sua resposta com base no resultado desejado

**2. Analisador**: um m√©todo para extrair a sa√≠da de texto do modelo em uma estrutura desejada (geralmente json)

````python
from langchain.output_parsers import ResponseSchema, StructuredOutputParser
````

````python
# Como voc√™ gostaria que sua resposta fosse estruturada. Este √© basicamente um modelo de prompt sofisticado
response_schemas = [
    ResponseSchema(
        name="bad_string", description="Esta √© uma string de entrada do usu√°rio mal formatada"
    ),
    ResponseSchema(
        name="good_string", description="Esta √© a sua resposta, uma resposta reformatada"
    ),
]

# How you would like to parse your output
output_parser = StructuredOutputParser.from_response_schemas(response_schemas)
````

````python
# See the prompt template you created for formatting
format_instructions = output_parser.get_format_instructions()
print(format_instructions)
````

````python
template = """
Voc√™ receber√° uma string mal formatada de um usu√°rio.
Reformate-o e certifique-se de que todas as palavras estejam escritas corretamente, incluindo nomes de pa√≠ses, cidades e estados. 

{format_instructions}

% ENTRADA DO USU√ÑRIO:
{user_input}

SUA RESPOSTA:
"""

prompt = PromptTemplate(
    input_variables=["user_input"],
    partial_variables={"format_instructions": format_instructions},
    template=template,
)

promptValue = prompt.format(user_input="Be vind a S√£o Pa!")

print(promptValue)
````

````python
llm_output = llm(promptValue)
llm_output
````

````python
output_parser.parse(llm_output)
````

### 2. Summarization Chain (Sumariza√ß√£o)

[Summarization Chain](https://python.langchain.com/docs/modules/chains/popular/summarize) percorre facilmente v√°rios documentos e obt√©m um resumo. Existem v√°rios tipos de cadeia, como Stuffing, Map-Reduce, Refine, Map-Rerank. Confira a [documenta√ß√£o](https://python.langchain.com/docs/modules/chains/how_to/) para outros tipos de cadeia al√©m de `refine`.

````python
wikipedia.set_lang('pt')
page = wikipedia.page('Mem√≥rias P√≥stumas de Br√°s Cubas')
with open('big.txt', 'w') as f:
    f.write(page.content)
````

````python
from langchain.chains.summarize import load_summarize_chain
from langchain.document_loaders import TextLoader
from langchain.text_splitter import CharacterTextSplitter

loader = TextLoader(
    'big.txt'
)
documents = loader.load()

# Prepare seu divisor
text_splitter = CharacterTextSplitter(chunk_size=1500, chunk_overlap=0)

# Divida seus documentos em textos
texts = text_splitter.split_documents(documents)
````

````python
prompt_template = """Escreva um resumo conciso do texto a seguir:
{text}
RESUMO:"""
prompt = PromptTemplate.from_template(prompt_template)

refine_template = (
     "Seu trabalho √© produzir um resumo final\n"
     "Temos a oportunidade de refinar o resumo existente"
     "(somente se necess√°rio) com mais contexto abaixo.\n"
     "---------------\n"
     "{text}\n"
     "---------------\n"
     "Dado o novo contexto, refine o resumo original em portugu√™s"
     "Se o contexto n√£o for √∫til, retorne o resumo original."
)
refine_prompt = PromptTemplate.from_template(refine_template)

print(f"# de palavras no documento = {len(documents[0].page_content)}")

# H√° muita complexidade escondida nesta linha. Eu encorajo voc√™ a conferir o v√≠deo acima para mais detalhes
chain = load_summarize_chain(
    llm,
    chain_type="refine",
    question_prompt=prompt,
    refine_prompt=refine_prompt,
    return_intermediate_steps=True,
    input_key="input_documents",
    output_key="output_text",
)

result = chain({"input_documents": texts}, return_only_outputs=True)
````

````python
print(result["output_text"])
````
